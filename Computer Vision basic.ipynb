```python
import cv2
import numpy as np

title = 'mouse event'
img = cv2.imread("./Images/blank_500.jpg")
cv2.imshow(title,img)

def onMouse(event, x, y, flag, param):
    global img
    #print(event,x,y,)
    if event==cv2.EVENT_LBUTTONDOWN:
        randcol=255*(np.random.rand(3))
        randr=int(100*(np.random.rand(1)))
        cv2.circle(img,(x,y),randr,randcol,-1)
        cv2.imshow(title,img)
    elif event==cv2.EVENT_RBUTTONDOWN:
        img = cv2.imread("./Images/blank_500.jpg")
        cv2.imshow(title,img)
    elif event==cv2.EVENT_LBUTTONDBLCLK:
        randcol=255*(np.random.rand(3))
        randx=int(100*(np.random.rand(1)))
        randy=int(100*(np.random.rand(1)))
        cv2.rectangle(img,(x-randx,y-randy),(x+randx,y+randy),randcol,-1)
        cv2.imshow(title,img)

cv2.setMouseCallback(title, onMouse)

while True:
    if cv2.waitKey(30)&0xFF==27:
        break
    elif cv2.waitKey(30)&0xFF==113:
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

title='Practice_3'
file_path='./test/practice_3.jpg'
img=cv2.imread("./Images/blank_500.jpg")
cv2.imshow(title,img)
img_copy = img.copy()
while True:
    randcol=255*(np.random.rand(3))    #색상 랜덤
    randr=int(499*(np.random.rand(1))) #반지름 랜점
    (x1,y1)=499*((np.random.rand(2)))  #위치 랜덤 1
    (x2,y2)=499*(np.random.rand(2))    #위치 랜덤 2
    x1=int(x1); x2=int(x2); y1=int(y1); y2=int(y2) #랜덤 위치 어레이 내부 값을 int로
    randthick=(int(10*(np.random.rand(1)))+1)      #랜덤 두께
    if cv2.waitKey(0)&0xFF==108: #press l to draw line **only use small letter
        cv2.line(img_copy,(x1,y1),(x2,y2),randcol,randthick)
        cv2.imshow(title,img_copy)
    elif cv2.waitKey(0)&0xFF==99:  #press c to draw circle **only use small letter
        cv2.circle(img_copy,(x1,y1),randr,randcol,randthick)
        cv2.imshow(title,img_copy)
    elif cv2.waitKey(0)&0xFF==114: #press r to draw rectangle **only use small letter
        cv2.rectangle(img_copy,(x1,y1),(x2,y2),randcol,randthick)
        cv2.imshow(title,img_copy)
    elif cv2.waitKey(0)&0xFF==101: #press e to erase **only use small letter
        img_copy=img.copy()
        cv2.imshow(title,img_copy)
    elif cv2.waitKey(0)&0xFF==113: #press q to quit **only use small letter
        cv2.imwrite(file_path,img_copy)
        break
        
cv2.destroyAllWindows()
```

```python
import cv2

title = 'img' # 원본이미지 창 이름
title2= 'ROI' # ROI 설정한 부분 창 이름
roi_path='./ROI.jpg'#ROI 설정한 부분 저장할 위치
blue, red = (255,0,0),(0,0,255)
isDragging=False # 드래그 상태인지 아닌지 구분하는 변수 -> 이거 떠올리는거 못해서 힘들었습니다.
x0, y0= -1, -1 # 초기값 설정
img=cv2.imread("./Images/food.jpg") # 원본 이미지 어레이화
cv2.imshow(title,img) # 이미지 창 켜기

def onMouse(event, x, y, flag, param):
    global img, x0, y0, isDragging, blue, red, title, title2, roi_path # 변수 불러오가
    if event==cv2.EVENT_LBUTTONDOWN:
        isDragging =True # 드래그중입니다.
        x0=x # 초기 위치 설정
        y0=y # 초기 위치 설정
        cv2.destroyWindow("error")#에러창 닫기
        cv2.destroyWindow(title2)#기존 ROI창 닫기 -> 트라이 & 엑셉트 구문 넣으면 더 깔끔할것
    elif event==cv2.EVENT_MOUSEMOVE:
        if isDragging: # 드래그 중일때만 동작하는 If 문
            img_draw=img.copy() # 원본 이미지 보호용 카피 어레이 생성.
                                #마우스가 움직일때마다 원본이미지를 불러와서 초기화해야하는 중요한 부분.
            cv2.rectangle(img_draw,(x0,y0),(x,y),blue,2) #파란 사각형 그리기
            cv2.imshow(title,img_draw) # 그린거 바로 표시
    elif event==cv2.EVENT_LBUTTONUP:
        if isDragging: # 드래그 중일때만 동작.
            isDragging=False # 마우스 뗐으니 드래그 off
            img_draw=img.copy() # 원본이미지 보호용 카피 어레이 생성.
            if y>y0 and x>x0: # 초기위치보다 왼쪽 아래일때만 동작.
                cv2.rectangle(img_draw,(x0,y0),(x,y),red,2) # 붉은 사각형 그리기
                roi=img[y0:y,x0:x] # 사각형 그린 부분의 어레이 저장
                cv2.imshow(title,img_draw) # 사각형이 그려진 원본사진 띄우기
                cv2.imshow(title2,roi) # 사각형 그린 부분만 사진 새 창 띄우기
                cv2.imwrite(roi_path,roi) #ROI 지정부분 저장함.
                print("ROI setting is saved : Rectangle from (%d ,%d) to (%d,%d)"%(x0,y0,x,y)) # ROI 세팅 좌표 설정
            else:
                cv2.imshow(title,img_draw)
                error = cv2.imread("./Images/skull.jpg")
                cv2.imshow("error",error)
                print("에러! 좌상에서 우하로 드래그하세요.")

cv2.setMouseCallback(title, onMouse)

while True:
    if cv2.waitKey(30)&0xFF==27:# 아스키코트 정수 ESC = 27, ESC눌러서 종료
        break
    elif cv2.waitKey(30)&0xFF==ord('q'): #아스키코드 정수 q = 119, 소문자 q 눌러서 종료
        break
cv2.destroyAllWindows()
```
